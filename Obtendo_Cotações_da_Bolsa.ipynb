```python
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime

data_atual = datetime.now().strftime('%Y-%m-%d')

cotacao_weg = yf.download('WEGE3.SA', start='2023-01-01', end=data_atual)
cotacao_petr = yf.download('PETR4.SA', start='2023-01-01', end=data_atual)
cotacao_mglu = yf.download('MGLU3.SA', start='2023-01-01', end=data_atual)
cotacao_itsa = yf.download('ITSA4.SA', start='2023-01-01', end=data_atual)

fig, ax = plt.subplots(figsize=(17, 12))

cotacao_weg["Adj Close"].plot(ax=ax, label='WEG')
cotacao_petr["Adj Close"].plot(ax=ax, label='Petrobras')
cotacao_mglu["Adj Close"].plot(ax=ax, label='Magalu')
cotacao_itsa["Adj Close"].plot(ax=ax, label='Itausa')

ax.set_xlabel('Data')
ax.set_ylabel('Preço (R$)')
ax.set_title('Cotações WEG e Petrobras')

plt.legend()
plt.show()

```
